```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

```python
data_dir = "/kaggle/input/datasets/dimthkiuoanh/nih-chest-xray-data-preprocessing-output"
```

```python
train_dir = os.path.join(data_dir, "train_224/train")

normal_dir = os.path.join(train_dir, "NORMAL")
pneumonia_dir = os.path.join(train_dir, "PNEUMONIA")

print("NORMAL:", len(os.listdir(normal_dir)))
print("PNEUMONIA:", len(os.listdir(pneumonia_dir)))
```

```python
num_images = 5   # số ảnh mỗi lớp

plt.figure(figsize=(12,6))

# NORMAL
for i in range(num_images):
    img = random.choice(os.listdir(normal_dir))
    path = os.path.join(normal_dir, img)
    
    plt.subplot(2, num_images, i+1)
    plt.imshow(load_img(path))
    plt.title("NORMAL")
    plt.axis("off")

# PNEUMONIA
for i in range(num_images):
    img = random.choice(os.listdir(pneumonia_dir))
    path = os.path.join(pneumonia_dir, img)
    
    plt.subplot(2, num_images, num_images+i+1)
    plt.imshow(load_img(path))
    plt.title("PNEUMONIA")
    plt.axis("off")

plt.show()
```
