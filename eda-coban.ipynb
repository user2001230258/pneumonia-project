```python
import pandas as pd
import os
```

```python
base_path = "/kaggle/input/datasets/dimthkiuoanh/nih-chest-xray-data-preprocessing-output"

train_df = pd.read_csv(os.path.join(base_path, "split_train.csv"))
val_df   = pd.read_csv(os.path.join(base_path, "split_val.csv"))
test_df  = pd.read_csv(os.path.join(base_path, "split_test.csv"))
```

```python
def count_classes(df, name):
    print(f"\n{name}")
    print(df["label"].value_counts())

count_classes(train_df, "TRAIN")
count_classes(val_df, "VAL")
count_classes(test_df, "TEST")
```

```python
summary = pd.DataFrame({
    "Split": ["Train", "Val", "Test"],
    "NORMAL": [
        (train_df["label"] == "NORMAL").sum(),
        (val_df["label"] == "NORMAL").sum(),
        (test_df["label"] == "NORMAL").sum()
    ],
    "PNEUMONIA": [
        (train_df["label"] == "PNEUMONIA").sum(),
        (val_df["label"] == "PNEUMONIA").sum(),
        (test_df["label"] == "PNEUMONIA").sum()
    ]
})

summary["Total"] = summary["NORMAL"] + summary["PNEUMONIA"]
summary
```

```python
total = summary["Total"].sum()

summary["Ratio (%)"] = summary["Total"] / total * 100
summary
```
